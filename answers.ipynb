### Q1.9 — Concept (short answer)
In a markdown cell or comment, explain:
- What is the difference between a **NumPy view** and a **copy**?
- Why are NumPy operations faster than Python lists for numeric work?

**View vs copy:** A *view* shares the same underlying data buffer as the original array — changes to the view affect the original. A *copy* is independent (`arr.copy()`). Slicing often returns a view; fancy indexing usually returns a copy.

**Why NumPy is faster:** NumPy stores homogeneous data in contiguous C arrays and runs operations in compiled C code (vectorization), avoiding slow Python loops and per-element type checking.

### Q2.10 — Concept (short answer)
Explain the difference between:
- `loc` vs `iloc`
- `groupby().agg()` vs `groupby().transform()`

**`loc` vs `iloc`:** `loc` selects by **labels** (column names, index labels). `iloc` selects by **integer position** (0-based row/column numbers).

**`agg()` vs `transform()`:** `agg()` collapses each group to a single summary value (e.g. mean per group). `transform()` returns a result **the same length as the original DataFrame**, broadcasting the group statistic back to every row (e.g. subtract each row's value from its group mean).

### Q5.4 — IQR outlier rule
Using the **1.5 × IQR** rule on `Revenue`, find lower and upper fences and identify outliers.

```python
revenue = sales_df['Revenue']
q1, q3 = revenue.quantile(0.25), revenue.quantile(0.75)
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

outliers = sales_df[(revenue < lower_fence) | (revenue > upper_fence)]

print(f"Q1={q1}, Q3={q3}, IQR={iqr}")
print(f"Lower fence: {lower_fence:.2f}")
print(f"Upper fence: {upper_fence:.2f}")
display(outliers)
```

### Q5.7 — Hypothesis testing setup
**Scenario:** Is average `Revenue` for `Phone` orders different from `Laptop` orders?

1. State null and alternative hypotheses (H₀, H₁)
2. Run an independent two-sample t-test (`scipy.stats.ttest_ind`)
3. Interpret the p-value at α = 0.05

```python
phone_rev = sales_df.loc[sales_df['Product'] == 'Phone', 'Revenue']
laptop_rev = sales_df.loc[sales_df['Product'] == 'Laptop', 'Revenue']

print("H₀: Mean revenue for Phone = Mean revenue for Laptop")
print("H₁: Mean revenue for Phone ≠ Mean revenue for Laptop")

t_stat, p_value = stats.ttest_ind(phone_rev, laptop_rev, equal_var=False)
print(f"\nPhone mean: ${phone_rev.mean():,.2f}")
print(f"Laptop mean: ${laptop_rev.mean():,.2f}")
print(f"t-statistic: {t_stat:.3f}, p-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject H₀ at α=0.05 — evidence of a difference in mean revenue.")
else:
    print("Fail to reject H₀ at α=0.05 — no significant difference detected.")
```

### Q5.10 — Concept (short answer)
Define **population** vs **sample**, and explain why we use **sample standard deviation** (ddof=1) when estimating population SD from data.

**Population** is the entire group you want to study; a **sample** is a subset used to estimate population characteristics.

We use **ddof=1** (divide by n−1) in sample standard deviation because we lose one degree of freedom when estimating the mean from the same data. This corrects bias and gives an unbiased estimate of the population variance.

### Q6.2 — Normal distribution
1. Generate 5,000 samples from Normal(μ=100, σ=15)
2. Overlay the theoretical PDF on a histogram
3. What percentage of values fall within 1 standard deviation of the mean? (empirical vs ~68%)

```python
mu, sigma = 100, 15
samples = np.random.normal(mu, sigma, size=5000)
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 200)
pdf = stats.norm.pdf(x, mu, sigma)

within_1sd = np.mean((samples >= mu - sigma) & (samples <= mu + sigma))

plt.figure(figsize=(8, 4))
plt.hist(samples, bins=30, density=True, alpha=0.7, edgecolor='black', label='Samples')
plt.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')
plt.title('Normal(100, 15) Distribution')
plt.legend()
plt.tight_layout()
plt.show()

print(f"Empirical % within 1 SD: {within_1sd * 100:.1f}% (theoretical ≈ 68%)")
```

### Q6.4 — Binomial distribution

### Q6.7 — Central Limit Theorem
1. Draw 10,000 samples, each of size n = 30, from Exponential(λ=2)
2. Compute the mean of each sample → 10,000 sample means
3. Plot the distribution of sample means and compare to Normal(μ, σ/√n)

```python
lam = 2
n = 30
n_samples = 10000

samples = np.random.exponential(scale=1 / lam, size=(n_samples, n))
sample_means = samples.mean(axis=1)

pop_mean = 1 / lam
pop_std = 1 / lam
se = pop_std / np.sqrt(n)

plt.figure(figsize=(8, 4))
plt.hist(sample_means, bins=40, density=True, alpha=0.7, edgecolor='black', label='Sample means')
x = np.linspace(sample_means.min(), sample_means.max(), 200)
plt.plot(x, stats.norm.pdf(x, pop_mean, se), 'r-', linewidth=2, label='Normal approx.')
plt.title('Central Limit Theorem: Means of Exponential(λ=2) samples')
plt.legend()
plt.tight_layout()
plt.show()
```

### Q6.10 — Monte Carlo simulation
**Scenario:** Roll two fair six-sided dice 50,000 times.

1. Estimate P(sum = 7)
2. Plot the distribution of sums
3. Compare your estimate to the exact probability 6/36

```python
n_rolls = 50_000
die1 = np.random.randint(1, 7, size=n_rolls)
die2 = np.random.randint(1, 7, size=n_rolls)
sums = die1 + die2

p_sum_7 = np.mean(sums == 7)
exact_p = 6 / 36

plt.figure(figsize=(8, 4))
plt.hist(sums, bins=np.arange(1.5, 13.5, 1), edgecolor='black', density=True)
plt.title('Distribution of Two-Dice Sums (50,000 rolls)')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.xticks(range(2, 13))
plt.tight_layout()
plt.show()

print(f"Estimated P(sum=7): {p_sum_7:.4f}")
print(f"Exact probability:  {exact_p:.4f}")
```

### B1. End-to-end analysis
Using only this notebook's tools:
1. Clean and enrich `sales_df` (date features, revenue per unit)
2. Produce 3 Seaborn/Matplotlib charts that answer a business question you define
3. Report one descriptive stat, one hypothesis test, and one probability calculation relevant to your question

```python
# Business question: Which region and product drive the most revenue?

df = sales_df.copy()
df['Revenue_Per_Unit'] = df['Revenue'] / df['Units_Sold']
df['Month'] = df['Date'].dt.month_name()

region_totals = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
product_totals = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

sns.barplot(x=region_totals.index, y=region_totals.values, ax=axes[0], palette='Blues_d')
axes[0].set_title('Total Revenue by Region')
axes[0].set_ylabel('Revenue ($)')

sns.barplot(x=product_totals.index, y=product_totals.values, ax=axes[1], palette='Greens_d')
axes[1].set_title('Total Revenue by Product')
axes[1].set_ylabel('Revenue ($)')

sns.boxplot(data=df, x='Product', y='Revenue_Per_Unit', ax=axes[2])
axes[2].set_title('Revenue per Unit by Product')

plt.tight_layout()
plt.show()

# Descriptive stat
print(f"Top region: {region_totals.idxmax()} (${region_totals.max():,.0f})")
print(f"Mean revenue per unit: ${df['Revenue_Per_Unit'].mean():.2f}")

# Hypothesis test: Phone vs Laptop revenue
phone = df.loc[df['Product'] == 'Phone', 'Revenue']
laptop = df.loc[df['Product'] == 'Laptop', 'Revenue']
_, p_val = stats.ttest_ind(phone, laptop, equal_var=False)
print(f"Phone vs Laptop t-test p-value: {p_val:.4f}")

# Probability: if each call has 30% close rate over 10 calls
p_at_least_3 = 1 - stats.binom.cdf(2, 10, 0.3)
print(f"P(≥3 deals in 10 calls at p=0.3): {p_at_least_3:.3f}")
```

**Summary:** East and South lead in total revenue, while **Phone** is the top product by revenue. Revenue per unit is highest for **Phone** and **Laptop**, reflecting their higher price points. A t-test shows no statistically significant difference between Phone and Laptop mean order revenue at α = 0.05 (small sample of 5 orders each). If a rep has a 30% close rate on 10 calls, there is about a 62% chance of closing at least 3 deals — useful for setting realistic sales targets.
